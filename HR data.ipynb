```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('cleaned_HR_dataset.csv')
```

```python

import plotly.graph_objects as go

# Calculate average salary
average_salary = df['Salary'].mean()

# Display KPI using Plotly
fig = go.Figure(go.Indicator(
    mode="gauge+number",
    value=average_salary,
    title={'text': "Average Salary"},
    gauge={
        'axis': {'range': [0, df['Salary'].max()]},
        'bar': {'color': "darkblue"},
        'steps': [
            {'range': [0, average_salary], 'color': "lightblue"},
            {'range': [average_salary, df['Salary'].max()], 'color': "lightgray"}
        ]
    },
    domain={'x': [0, 1], 'y': [0, 1]}
))

fig.show()
```

```python
# Calculate average engagement score
average_engagement_score = df['EngagementSurvey'].mean()

# Display KPI using Matplotlib
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.text(0.5, 0.5, f'Average Engagement Score\n{average_engagement_score:.2f}', 
        fontsize=20, ha='center', va='center')
ax.set_title('Employee Engagement Score')
ax.axis('off')

plt.show()
```

```python
# Calculate percentage of still employed
total_employees = len(df)
still_employed = len(df[df['EmploymentStatus'] == 'N/A-StillEmployed'])
percentage_still_employed = (still_employed / total_employees) * 100

# Display KPI using Plotly
fig = go.Figure(go.Indicator(
    mode="gauge+number",
    value=percentage_still_employed,
    title={'text': "Percentage Still Employed"},
    gauge={
        'axis': {'range': [0, 100]},
        'bar': {'color': "darkgreen"},
        'steps': [
            {'range': [0, percentage_still_employed], 'color': "lightgreen"},
            {'range': [percentage_still_employed, 100], 'color': "lightgray"}
        ]
    },
    domain={'x': [0, 1], 'y': [0, 1]}
))

fig.show()
```

```python
import matplotlib.pyplot as plt


# Define the mapping for PerformanceScore
performance_mapping = {
    'Needs Improvement': 1,
    'Fully Meets': 2,
    'Exceeds': 3,
    'PIP': 0  # Assuming 'PIP' should be mapped to 0; adjust if necessary
}

# Replace categorical values with numeric values
df['PerformanceScore'] = df['PerformanceScore'].map(performance_mapping)

# Calculate average performance score
average_performance_score = df['PerformanceScore'].mean()

# Display KPI using Matplotlib
fig, ax = plt.subplots()
ax.text(0.5, 0.5, f'Average Performance Score\n{average_performance_score:.2f}', 
        fontsize=20, ha='center', va='center')
ax.set_title('Average Performance Score')
ax.axis('off')

plt.show()
```

```python
# Calculate total absences
total_absences = df['Absences'].sum()

# Display KPI using Plotly
fig = go.Figure(go.Indicator(
    mode="gauge+number",
    value=total_absences,
    title={'text': "Total Absences"},
    gauge={
        'axis': {'range': [0, df['Absences'].max()]},
        'bar': {'color': "darkred"},
        'steps': [
            {'range': [0, total_absences], 'color': "lightcoral"},
            {'range': [total_absences, df['Absences'].max()], 'color': "lightgray"}
        ]
    },
    domain={'x': [0, 1], 'y': [0, 1]}
))

fig.show()
```

```python
from datetime import datetime

# Calculate tenure
df['DateofHire'] = pd.to_datetime(df['DateofHire'], format='%d-%m-%Y')
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'], format='%d-%m-%Y', errors='coerce')
df['Tenure'] = (df['DateofTermination'].fillna(datetime.today()) - df['DateofHire']).dt.days / 365

# Tenure distribution
plt.figure(figsize=(12, 6))
sns.histplot(df['Tenure'], bins=20, kde=True)
plt.title('Distribution of Employee Tenure')
plt.xlabel('Tenure (Years)')
plt.ylabel('Frequency')
plt.show()

# Tenure vs Performance Score
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Tenure', y='PerformanceScore', data=df)
plt.title('Performance Score vs Tenure')
plt.xlabel('Tenure (Years)')
plt.ylabel('Performance Score')
plt.show()
```
